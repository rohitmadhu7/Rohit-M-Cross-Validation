```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('titanic_dataset.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data = data.drop(columns='Cabin', axis=1)
```

```python
data['Age'].fillna(data['Age'].mean(), inplace=True)
```

```python
print(data['Embarked'].mode())
```

```python
print(data['Embarked'].mode()[0])
```

```python
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data['Survived'].value_counts()
```

```python
data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
data.head()
```

```python
X = data.drop(columns = ['PassengerId','Name','Ticket','Survived'],axis=1)
Y = data['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

SVM

```python
from sklearn.svm import SVC
svmclf= SVC(kernel= 'linear')
svmclf.fit (X_train,Y_train)
```

```python
Y_pred_svm= svmclf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('Accuracy = ',accuracy_score(Y_test,Y_pred_svm))
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn.fit(X_train, Y_train)
```

```python
Y_pred = knn.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('Accuracy = ',accuracy_score(Y_test,Y_pred))
```

k fold cross validation for knn

```python
from sklearn.model_selection import KFold
knn_model=KNeighborsClassifier()
```

```python
kfold_validator=KFold(10)
```

```python
for train_index,test_index in kfold_validator.split(X,Y):
  print('Training index :',train_index)
  print('Testinging index :',test_index)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score=cross_val_score(knn_model,X,Y,cv=kfold_validator)
```

```python
cv_score
```

```python
np.mean(cv_score)
```

stratified k-fold cross validation for knn

```python
from sklearn.model_selection import StratifiedKFold
```

```python
sk_fold_validator=StratifiedKFold(n_splits=10)
```

```python
for train_index,test_index in kfold_validator.split(X,Y):
  print('Training index :',train_index)
  print('Testinging index :',test_index)
```

```python
skcv_score=cross_val_score(knn_model,X,Y,cv=sk_fold_validator)
```

```python
skcv_score
```

```python
np.mean(skcv_score)
```
